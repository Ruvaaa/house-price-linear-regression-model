# This is our third attemp at making a linear regression model with Zillow to20 states data. 

### Imports

```python
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Data loading

```python
data = pd.read_csv("zillow_top20_states.csv")
data.head()
```

### Encode columns

```python
#encoder
le = LabelEncoder()

#columns to encode
to_encode = ['state_code','state_name','address','street','city','status','home_type']

#for loop to loop through columns
for col in to_encode:
    data[col] = le.fit_transform(data[col]) + 1
```

```python
data.head()
```

### drop unnecessary columns

```python
#drop columnsd
data.drop(columns = ['price_formatted','detail_url','has_open_house','open_house_date'], inplace = True)
```
